### Part 1: Building the base Anagram Checker

Given two valid strings, check to see if they are anagrams of each other. If it is, return `True`, else `False`. For this part, we can assume that uppercase letters are the same as if it was a lowercase character.

Examples of anagrams:
* Silent and Listen
* Night and Thing

Example outputs:
```python
anagram_checker("Silent", "listen") # True
anagram_checker("Silent", "Night") # False
anagram_checker("night", "Thing") # True
```

````python
# For testing purposes, we will write our code in the function
def anagram_checker(word_a, word_b):
  # Your code here
  """A function that checks if two words are anagrams of each other. 
  An anagram is a word formed by rearranging the letters of another word, using all the original letters exactly once. 
  The function should return True if the two words are anagrams, and False otherwise.
  
  Parameters:
  word_a (str): The first word to check.
  word_b (str): The second word to check.
  
  Rationale:
  To determine if two words are anagrams, we can follow these steps:
  1. Convert both words to lowercase to ensure the comparison is case-insensitive.
  2. Sort the letters of each word alphabetically. 
  3. Use comparison to check if the sorted letters of both words are the same.
  If the sorted letters of both words are the same, then the words are anagrams of each other.

  Returns:
  True (bool) if the two words are anagrams, False (bool) otherwise."""
  
  #turn all letters to lowercase
  word_a = word_a.lower() # converts all letters in word_a to lowercase
  word_b = word_b.lower() # converts all letters in word_b to lowercase

  #sort the letters in each word
  sorted_a = sorted(word_a) # returns a list of letters in word_a, sorted in alphabetical order (repeating letters will appear multiple times)
  sorted_b = sorted(word_b) # returns a list of letters in word_b, sorted in alphabetical order (repeating letters will appear multiple times)

  #check if the sorted letters are the same and return the result
  return sorted_a == sorted_b # returns True if sorted_a and sorted_b are the same list of letters, False otherwise

# Run your code to check using the words below:
anagram_checker("Silent", "listen")
````

````python
anagram_checker("Silent", "Night")
````

````python
anagram_checker("night", "Thing")
````

### Part 2: Expanding the functionality of the Anagram Checker

Using your existing and functional anagram checker, let's add a boolean option called `is_case_sensitive`, which will return `True` or `False` based on if the two compared words are anagrams and if we are checking for case sensitivity.

````python
def anagram_checker(word_a, word_b, is_case_sensitive):
  # Modify your existing code here
  """A function that checks if two words are anagrams of each other. 
  An anagram is a word formed by rearranging the letters of another word, using all the original letters exactly once. 
  The function should return True if the two words are anagrams, and False otherwise.
  
  Parameters:
  word_a (str): The first word to check.
  word_b (str): The second word to check.
  
  Rationale:
  To determine if two words are anagrams, we can follow these steps:
  1. Convert both words to lowercase to ensure the comparison is case-insensitive.
    1-1. If the comparison is case-sensitive, we can skip this step and keep the original case of the letters.
  2. Sort the letters of each word alphabetically. 
  3. Use comparison to check if the sorted letters of both words are the same.
  If the sorted letters of both words are the same, then the words are anagrams of each other.

  Returns:
  True (bool) if the two words are anagrams, False (bool) otherwise."""

  if is_case_sensitive: # if case sensitive
    #sort the letters in each word
    sorted_a = sorted(word_a) # returns a list of letters in word_a, sorted in alphabetical order (repeating letters will appear multiple times)
    sorted_b = sorted(word_b) # returns a list of letters in word_b, sorted in alphabetical order (repeating letters will appear multiple times)
  
  else: # if case insensitive
    #turn all letters to lowercase
    word_a = word_a.lower() # converts all letters in word_a to lowercase
    word_b = word_b.lower() # converts all letters in word_b to lowercase

    #sort the letters in each word
    sorted_a = sorted(word_a) # returns a list of letters in word_a, sorted in alphabetical order (repeating letters will appear multiple times)
    sorted_b = sorted(word_b) # returns a list of letters in word_b, sorted in alphabetical order (repeating letters will appear multiple times)

  #check if the sorted letters are the same and return the result
  return sorted_a == sorted_b # returns True if sorted_a and sorted_b are the same list of letters, False otherwise

# Run your code to check using the words below:
anagram_checker("Silent", "listen", False) # True
````
